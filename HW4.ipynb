```python
 ### Question 4




 #Step 1 Define the Objective Function and its derivatives,
 #Implement Newtons Method and Gradient Descent Method

import numpy as np
import matplotlib.pyplot as plt

def f(x):
    """Objective function."""
    x1, x2 = x
    return x1**2 + x2**2 - 6*x1 - 2*x2 + 10

def grad_f(x):
    """Gradient of the objective function."""
    x1, x2 = x
    return np.array([2*x1 - 6, 2*x2 - 2])

def hessian_f(x):
    """Hessian of the objective function."""
    return np.array([[2, 0], [0, 2]])

def newtons_method(f, grad_f, hessian_f, x0, n=10):
    """Newton's method for finding the function's minimizer, capturing objective function values."""
    x = x0
    values = [f(x)]
    for _ in range(n):
        grad = grad_f(x)
        hess = hessian_f(x)
        delta_x = np.linalg.solve(hess, -grad)
        x += delta_x
        values.append(f(x))
    return x, values

def gradient_descent(f, grad_f, x0, step_size=0.01, n=100):
    """Gradient descent method for finding the function's minimizer, capturing objective function values."""
    x = x0
    values = [f(x)]
    for _ in range(n):
        grad = grad_f(x)
        x -= step_size * grad
        values.append(f(x))
    return x, values




#Perform Optimization and Plot REsults

x0 = np.array([1, 1], dtype=float) 


minimizer_newton, values_newton = newtons_method(f, grad_f, hessian_f, x0)
minimizer_gradient_descent, values_gradient_descent = gradient_descent(f, grad_f, x0, step_size=0.01, n=100)

print(minimizer_gradient_descent)
print(minimizer_newton)
# Plot
plt.figure(figsize=(10, 6))
plt.plot(values_newton, '-o', label="Newton's Method")
plt.plot(values_gradient_descent, '-x', label="Gradient Descent")
plt.xlabel('Iteration')
plt.ylabel('Objective Function Value')
plt.title('Objective Function Value Across Iterations')
plt.legend()
plt.show()



```

```python
#Question 2
import numpy as np
import matplotlib.pyplot as plt

# Define the lines based on the inequalities
# For the cashews inequality: x/2 + y/3 = 130 -> y = 390 - 3/2 x
# For the peanuts inequality: x/2 + 2y/3 = 170 -> y = 255 - 3/4 x

# Create an array of x values
x_values = np.linspace(0, 300, 400)

# Create the y values based on the equations above
y_cashews = 390 - 1.5 * x_values
y_peanuts = 255 - 0.75 * x_values

# Set up the plot
plt.figure(figsize=(10, 10))

# Plot the lines
plt.plot(x_values, y_cashews, label=r'$\frac{x}{2} + \frac{y}{3} = 130$ (Cashews Constraint)')
plt.plot(x_values, y_peanuts, label=r'$\frac{x}{2} + \frac{2y}{3} = 170$ (Peanuts Constraint)')

# Fill the feasible region
plt.fill_between(x_values, np.maximum(0, y_peanuts), np.minimum(y_cashews, 390), where=(y_peanuts >= 0), color='gray', alpha=0.5)

# Add labels and legend
plt.xlabel('Pounds of Premium Mixture (x)')
plt.ylabel('Pounds of Value Mixture (y)')
plt.title('Feasible Region for Nut Mixtures')
plt.legend()

# Add constraints
plt.axhline(0, color='black', lw=2)
plt.axvline(0, color='black', lw=2)

# Set the limits
plt.xlim(0, 300)
plt.ylim(0, 400)

# Show the plot
plt.show()
```
